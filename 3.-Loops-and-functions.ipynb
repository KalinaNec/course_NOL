# Loops, comprehensions and functions

### 1. Loops

A loop is a programming construct for repeated execution of a block of code. It is fundamental whenever the same operation must be applied to many values or objects.

Two common forms are used in Python. A `while` loop repeats as long as a condition remains true, whereas a `for` loop iterates over the elements of an iterable object such as a list, range, or dictionary.


````python
counter = 0
while counter < 10:
    print(counter)
    counter += 1
````

````python
counter = range(10)

for count in counter:
    print(count)
````

````python
animals = ["animal_1", "animal_2", "animal_3"]
colors = ["red", "blue", "green"]

for animal in animals:
    for color in colors:
        print(animal, color)
````

````python
animals = ["animal_1", "animal_2", "animal_3"]
colors = ["red", "blue", "green"]

for animal, color in zip(animals, colors):
    print(animal, color)
````

````python
neuron_dict = {
    "neuron_0": 2,
    "neuron_1": 15,
    "neuron_2": 8,
    "neuron_3": 22,
    "neuron_4": 5,
    "neuron_5": 11,
}

fast_spiking = {}

for neuron, firing_rate in neuron_dict.items():
    if firing_rate > 10:
        fast_spiking[neuron] = firing_rate
        print(f"{neuron} is a fast spiking neuron")
````

### 2. Comprehensions

A comprehension is a compact notation for constructing a new container from existing data. Python provides list, set, and dictionary comprehensions.

Conceptually, a comprehension often plays the role of a loop whose purpose is to build a new object. It is useful when the transformation is simple and can be expressed clearly in one line.


````python
my_list = [i.capitalize() for i in ["konrad", "Kalina", "adam", "mateusz", "mateusz"]]
my_list
````

````python
my_set = {i.capitalize() for i in ["konrad", "Kalina", "adam", "mateusz", "mateusz"]}
my_set
````

````python
my_dict = {
    name: affiliation
    for name, affiliation in zip(
        ["Konrad", "Kalina", "Mateusz"],
        ["Nencki", "Nencki", "Uni Heidelberg"],
    )
}
my_dict
````

````python
# This is not a comprehension - altough it follows the same syntax it is a generator expression - it provides a way to evaluate lazily - producing one value at a time.
# For instance when reading and processing large data (one file at a time to avoid OOM)
my_generator = (
    i.capitalize() for i in ["konrad", "Kalina", "adam", "mateusz", "mateusz"]
)
````

### 3. Functions

A function is a reusable block of code that performs a specific task. In general, a function accepts input through parameters, processes that input, and may return a result.

Functions are central in programming because they help structure code, reduce repetition, and make larger analyses easier to read, test, and reuse.


````python
import random

spikes_1 = random.sample(range(20, 300), 20)
spikes_2 = random.sample(range(20, 100), 20)
spikes_3 = random.sample(range(20, 88), 20)

brain_regions = ["parietal_cortex", "amygdala", "hypothalamus"]
per_nrn_spikes = [spikes_1, spikes_2, spikes_3]
````

This cell does not define a function yet. It only prepares example data that the later functions will use as input.

This is often worth stating explicitly during teaching, because participants may otherwise treat every code cell as doing the same kind of work. In practice, some cells prepare data, some define tools, and some run those tools.


````python
# Works but ugly and hard to understand
def combine_regions_and_n_spikes(brain_regions, per_nrn_spikes):
    output_dict = dict.fromkeys(brain_regions)
    for region, n_spikes in zip(brain_regions, per_nrn_spikes):
        output_dict[region] = n_spikes
    return output_dict
````

A common source of confusion is the `return` statement in a function definition cell. Even though the word `return` is present, nothing is shown below the cell yet.

The reason is that this cell only defines the function. The code inside the function body is not executed until the function is called later. When that later call happens, `return` sends the result back to the caller. It does not automatically print the result.


````python
data_dict = combine_regions_and_n_spikes(
    brain_regions,
    per_nrn_spikes,
)
data_dict
````

Here the function is actually called. The objects `brain_regions` and `per_nrn_spikes` are passed into the function as arguments.

Inside the function, the `for` loop matches each region to its spike list, stores the pairs in `output_dict`, and then `return output_dict` sends that finished dictionary back. The notebook shows the result because the last line of the cell is `data_dict`.


````python
# Good practices - type hints, short docstring, clear implementation
def combine_regions_and_n_spikes(
    brain_regions: list[str], per_nrn_spikes: list[int]
) -> dict[str, int]:
    """Auxfun to match number of spikes per nrn to brain regions"""
    return {region: number for region, number in zip(brain_regions, per_nrn_spikes)}
````

This version performs the same operation, but more compactly.

More explanation:

- `brain_regions: list[str]` is a type hint saying that `brain_regions` should be a list of strings,
- `-> dict[str, int]` describes the expected return type,
- the short text between triple quotes is a docstring,
- the dictionary comprehension replaces the explicit loop from the previous version.


````python
data_dict = combine_regions_and_n_spikes(
    brain_regions,
    per_nrn_spikes,
)
data_dict
````

````python
def calculate_mean_activity(per_nrn_spikes: list[int]):
    """Auxfun to calculate avg cell activity"""
    return sum(per_nrn_spikes) / len(per_nrn_spikes)
````

This function is a good place to explain what `return` actually gives back.

Step by step:

- `sum(per_nrn_spikes)` adds all values in the list,
- `len(per_nrn_spikes)` counts how many values there are,
- dividing one by the other gives the mean,
- `return` sends that single mean value back.

If a function finishes without a `return` statement, Python returns `None` by default.


````python
per_region_activity = {
    region: calculate_mean_activity(n_spikes) for region, n_spikes in data_dict.items()
}
per_region_activity
````

This cell can look scary because it combines two ideas: a dictionary comprehension and a function call.

The comprehension goes through each `(region, n_spikes)` pair in `data_dict.items()`. For every pair, `calculate_mean_activity(n_spikes)` is called, and the returned mean becomes the value stored under that region in the new dictionary.


### 4. Lambda functions (anonymous functions)

A lambda function is an anonymous function defined in a single expression. It provides a short way of writing very small functions without using a full `def` statement.

Lambda functions are often used together with functions such as `map()` and `filter()`. However, when the logic becomes more complex, a named function is usually preferable for readability.


````python
check = [10, 20, 40, 10, 15, 30, 105, 25, 85]

print("Lambda functions allow us to easily perform simple operations in one line without a full function definition.")
print("Map function is exactly what it sounds like: it maps our lambda function onto the provided object (list of integers):\n")
print(list(map(lambda x: x**2, check)), "\n")

print("Similarily if the lambda evaluates to booleans, we can apply a filter function to filter out specific values from our list:\n")
print(list(filter(lambda x: x > 20, check)))
````

````python
# Filter out regions with low activity
threshold = 70

# Difference between filter and map
print("Filtering returns the values that fulfill the condition\n")
print(dict(filter(lambda item: item[1] > threshold, per_region_activity.items())), "\n")

print("Mapping returns bool expression to tell us which indices fulfill the condition\n")
print(list(map(lambda item: item[1] > threshold, per_region_activity.items())))
````

````python
# Let's combine the analysis into a high level function
def analyze_brain_data(
    brain_regions: list[str],
    per_nrn_spikes: list[list[int]],
    filter_actvity: bool,
    activity_threshold: int | float | None = None,
) -> dict[str, int] | dict[None]:
    """Match brain regions to their neural activity and optionnaly perform filtering based on the average number of spikes

    Args:
        brain_regions: list of brain regions.
        per_nrn_spikes: list of lists containing number of spikes per neuron.
        filter_actvity: toggles whether to perform filtering based on cell spike count.
        activity_threshold: threshold for the activity if filter activity set.

    Returns:
        Dictionary of matched and optionally filtered average spike count in a brain region.
    """
    data_dict: dict = combine_regions_and_n_spikes(brain_regions, per_nrn_spikes)

    per_region_activity: dict = {
        region: calculate_mean_activity(n_spikes)
        for region, n_spikes in data_dict.items()
    }

    if filter_actvity and isinstance(activity_threshold, (int, float)):
        return dict(
            filter(lambda item: item[1] > activity_threshold, per_region_activity.items())
        )
    else:
        return per_region_activity
````

This final function combines several earlier steps into one larger procedure.

It is useful for explaining control flow:

- first it creates `data_dict`,
- then it computes `per_region_activity`,
- then it checks the `if` condition,
- finally it returns either the filtered dictionary or the full dictionary.

So a function does not have to return the same value in every situation. Different branches of the code can return different results.


````python
output = analyze_brain_data(brain_regions, per_nrn_spikes, True, 190)
````

### 5. In plain words

These ideas are often easier than they first look.

- A **loop** means: do the same action many times.
- A **comprehension** means: build a new collection in one compact line.
- A **function** means: pack a useful action into a reusable block.
- A **lambda** means: make a tiny throwaway function for a very small task.

In everyday terms:

- a loop is like checking every sample tube one by one,
- a comprehension is like making a cleaned-up list in one pass,
- a function is like writing a protocol once and reusing it,
- a lambda is like writing a very short one-line note for one quick operation.


````python
mouse_ids = ["mouse_01", "mouse_02", "mouse_03"]

print("Loop example:")
for mouse in mouse_ids:
    print(mouse, "was processed")

print("\nComprehension example:")
labels = [mouse.upper() for mouse in mouse_ids]
print(labels)
````

````python
def convert_ms_to_s(times_ms: list[int | float]) -> list[float]:
    """Convert times from milliseconds to seconds"""
    return [time / 1000 for time in times_ms]


reaction_times = [120, 250, 315, 98]
print(convert_ms_to_s(reaction_times))
````

### 7. More on lambda functions

A lambda function is just a very short function written in one line.

For example, these two things do the same job:

Regular function:

```python
def square(x):
    return x**2
```

Lambda function:

```python
lambda x: x**2
```

The lambda version is shorter, but also a bit harder to read when it becomes too complicated.

So in plain words:

- use `lambda` for teeny tiny simple operations,
- use a normal `def` function for bigger or clearer code.


````python
def add_one(x):
    return x + 1


numbers = [1, 2, 3, 4]

print("Using a normal function:")
print(list(map(add_one, numbers)))

print("\nUsing lambda:")
print(list(map(lambda x: x + 1, numbers)))
````

````python
cell_counts = [12, 4, 19, 7, 25, 3]

print("Keep only larger values:")
print(list(filter(lambda x: x > 10, cell_counts)))

print("Convert counts to squared values:")
print(list(map(lambda x: x**2, cell_counts)))
````
