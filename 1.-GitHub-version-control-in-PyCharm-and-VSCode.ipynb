# GitHub version control in PyCharm and VS Code


### 1. What is version control?

Version control is a way of saving the history of a project. Instead of having files like  `analysis_final.py`, `analysis_final_really_final.py` and `analysis_final_fixed.py`, we keep one project folder and record snapshots of changes over time.

In Git, these snapshots are called **commits**. Each commit has a message describing what was changed.

GitHub is a website that can store Git repositories online. This means your project can live both on your computer and on the internet, which makes backup, collaboration and sharing much easier.


### 2. Why do we use it?

Version control is useful because it lets us:

- go back to older versions of files,
- see exactly what changed,
- understand who changed what,
- collaborate without sending code by email,
- keep an online backup of our work,
- work on the same project from more than one computer.

Even when working alone, version control is very useful. It protects us from accidental mistakes, broken code, overwritten files and forgotten analysis steps.


### 3. Why is this useful in biology?

Biology projects often involve scripts for image analysis, behavioural analysis, statistics, microscopy pipelines, sequencing workflows or figure generation. These projects change all the time.

A few common examples:

- you change a threshold in an image analysis pipeline and want to know when that happened,
- you update a notebook that generates a figure for a paper,
- your supervisor or collaborator wants to see the latest version of the code,
- several students work on one repository,
- you come back to the project after three months and need to understand what you did.

Version control helps reproducibility. If somebody asks how a result was produced, a repository history gives a much clearer answer than a folder full of renamed files.


### 4. Example: project history without and with Git

Below is a toy example. It shows how a project history often looks without version control, and how the same thing looks with version control.


````python
without_git = [
    "cell_counting.py",
    "cell_counting_final.py",
    "cell_counting_final2.py",
    "cell_counting_final2_REAL.py",
]

with_git = [
    "Initial version of cell counting pipeline",
    "Fix thresholding bug in hippocampus images",
    "Add plot of per-animal counts",
    "Update README with analysis steps",
]

print("Without Git:")
for name in without_git:
    print(" -", name)

print("\nWith Git commit history:")
for i, commit in enumerate(with_git, start=1):
    print(f" {i}. {commit}")
````

### 5. Basic words

- **Repository (repo)**: a project folder tracked by Git.
- **Commit**: a saved snapshot of changes.
- **Push**: send your local commits to GitHub.
- **Pull**: bring changes from GitHub to your computer.
- **Clone**: download an existing GitHub repository to your computer.
- **History / Log**: the list of commits.
- **Diff**: the exact lines that changed.
- **Branch**: a parallel line of work. For now, we can stay on the main branch.


### 6. Example: what is a commit?

A commit is easiest to understand as a labeled snapshot in the project history.


````python
project_history = [
    {"commit": 1, "message": "Create notebook for neuron analysis"},
    {"commit": 2, "message": "Add preprocessing for calcium traces"},
    {"commit": 3, "message": "Fix wrong axis label in final plot"},
]

for entry in project_history:
    print(f"Commit {entry['commit']}: {entry['message']}")
````

For this course, the important point is that `uv` helps keep each project's Python environment separate and reproducible.

### 15. Cloning an existing repository

**Cloning** means copying a repository from GitHub to your computer.

This is what you do when:

- the repository already exists on GitHub,
- you want to join an existing project,
- you want a copy on a new computer.

Typical terminal command:

```bash
git clone https://github.com/USERNAME/REPOSITORY.git
cd REPOSITORY
```

You can also clone graphically:

- in **VS Code**, use the Command Palette or Source Control tools,
- in **PyCharm**, use **Get from VCS** on the start screen.


### 16. Connecting VS Code to GitHub

In VS Code:

1. Open VS Code.
2. Open your project folder or clone a repository.
3. Click the **Source Control** icon on the left sidebar.
4. If VS Code asks you to sign in, choose GitHub login in the browser.
5. Approve the connection.

After that, VS Code can publish a local repository to GitHub and synchronize changes using the graphical interface.


### 17. Connecting PyCharm to GitHub

In PyCharm, the cleanest path is usually browser login.

1. Open PyCharm.
2. Open a project folder.
3. Go to **Git** or **VCS** in the top menu.
4. If Git is not enabled yet, choose **Enable Version Control Integration** and select **Git**.
5. Open the GitHub account settings.
6. Choose login through GitHub in the browser.
7. Approve the connection.

After that, PyCharm can connect your local project to GitHub and use buttons for commit, push, pull, and history.


### 18. Starting a repository from a local project folder

This is a very common situation: you already have a folder with Python files or notebooks and now want to put it on GitHub.

General idea:

1. Open the project folder in the editor.
2. Turn on Git for that folder.
3. Add a `.gitignore` file.
4. Add a `pyproject.toml` file if this is a Python project.
5. Review the files that Git wants to track.
6. Make the first commit.
7. Publish or share the project to GitHub.


### 19. What is `.gitignore` and why is it there?

`.gitignore` tells Git which files or folders should **not** be tracked.

Why this matters:

- some files are generated automatically,
- some files are local to your computer,
- some files are temporary,
- some files are too large or noisy,
- some files may contain secrets.

Typical examples to ignore in a Python project:

- `__pycache__/`
- `.ipynb_checkpoints/`
- `.venv/`
- `.env`
- `.DS_Store`

A simple example:

```gitignore
__pycache__/
.ipynb_checkpoints/
.venv/
.env
.DS_Store
```

Without `.gitignore`, people often commit many useless autogenerated files by accident.


### 20. What is `pyproject.toml` and why is it there?

`pyproject.toml` is a standard Python project configuration file.

Depending on the project, it can describe:

- the project name,
- the Python version requirement,
- dependencies,
- build settings,
- tool configuration.

Why it is useful:

- it gives the project one clear configuration file,
- tools such as `uv` understand it,
- it makes environments more reproducible,
- it helps other people understand how to set up the project.

A minimal example might look like this:

```toml
[project]
name = "python-beginner-project"
version = "0.1.0"
description = "First practice repository for learning Git and GitHub"
requires-python = ">=3.11"
dependencies = []
```

You do not always need a whole `pyproject.toml`, but having one early is a good practice for Python projects.


### 21. Example project layout

A beginner Python repository might look like this:

```text
python-beginner-project/
  .gitignore
  pyproject.toml
  README.md
  src/
    example.py
  notebooks/
    first_analysis.ipynb
```

What these files do:

- `README.md` explains the project.
- `.gitignore` keeps junk and local machine files out of Git.
- `pyproject.toml` describes the Python project and dependencies.
- `src/` contains code.
- `notebooks/` contains notebooks.


### 22. First commit and first push

Once the repository is initialized and your files look correct, the basic workflow is:

```bash
git add .
git commit -m "Initial commit"
git push -u origin main
```

What each line means:

- `git add .` prepares the current changes.
- `git commit -m ...` saves a snapshot locally.
- `git push -u origin main` uploads the commits to GitHub.

If you use the editor GUI instead of terminal, the same ideas still apply: review files, commit with a clear message, then push.


### 23. First commit and push in PyCharm

A simple GUI-only workflow in PyCharm looks like this:

1. Open the project.
2. Enable Git version control if it is not enabled yet.
3. Open the **Commit** window.

5. Write a commit message, for example: `Add .gitignore and pyproject.toml`.

A good commit message should say what changed, not only that something changed.


### 24. First commit and push in VS Code

In VS Code:

1. Open the project folder.
2. Open **Source Control**.
3. If needed, click **Initialize Repository**.
4. Changed files will appear in the Source Control panel.
5. Stage the files you want to include.
6. Write a commit message.
7. Click **Commit**.

If VS Code asks you to sign in to GitHub, do that in the browser and then continue.


### 25. Good commit messages

A good commit message says what changed.

Good examples:

- `Add initial project structure`
- `Add .gitignore and pyproject.toml`
- `Connect notebook to uv environment`
- `Fix typo in README`

Not so good examples:

- `changes`
- `fix`
- `stuff`


### 26. What is the difference between clone, commit, push, and pull?

- **Clone**: download an existing repository from GitHub to your computer.
- **Commit**: save a snapshot on your computer.
- **Push**: send your local commits to GitHub.
- **Pull**: download newer changes from GitHub to your computer.


If you commit but do not push, your work is saved locally but is not yet on GitHub.

If somebody else pushed changes to GitHub, you need to pull them before your local project is fully up to date.


### 27. Example: local history and remote history

This example shows the idea behind local commits and pushed commits.


````python
local_commits = [
    "Initial notebook",
    "Add preprocessing section",
    "Fix typo in results description",
]

github_commits = [
    "Initial notebook",
    "Add preprocessing section",
]

print("Local history:")
for commit in local_commits:
    print(" -", commit)

print("\nGitHub history:")
for commit in github_commits:
    print(" -", commit)

print("\nThe last local commit is not yet on GitHub, so we still need to push.")
````

### 28. How to check what happened after a push

There are several easy ways.

On GitHub website:

1. Open your repository in the browser.
2. Look at the file list and latest commit message.
3. Click the commit history to see all commits.
4. Click a commit to see the exact lines added and removed.

In PyCharm:

1. Open the **Git Log** or **History** view.
2. Click a commit.
3. Inspect the diff to see what changed.

In VS Code:

1. Open **Source Control**.
2. Open the repository history or timeline view.
3. Click a commit.
4. Inspect the changed files and diff.


````python
import difflib

old_version = "threshold = 0.5\nprint('count cells')\n"
new_version = "threshold = 0.7\nprint('count cells')\nprint('save figure')\n"

diff = difflib.ndiff(old_version.splitlines(), new_version.splitlines())

for line in diff:
    print(line)
````

### 29. Good practical habits

- Pull before starting work on a shared repository.
- Commit often.
- Push regularly.
- Use clear commit messages.
- Check changed files before committing.
- Do not upload passwords, tokens or private keys.
- Do not upload large raw datasets to a normal GitHub repository.
- Keep `.gitignore`, `README.md`, and `pyproject.toml` under version control.

GitHub is excellent for code, text files, notebooks and small project files. It is usually not a good place for huge raw datasets or large binary files.


### 30. Minimal workflow to remember

If you forget everything else, remember this order:

1. Clone the repository or open your project folder.
2. Pull if the project already exists online.
3. Make your changes.
4. Review changed files.
5. Commit with a good message.
6. Push.
7. Check GitHub to confirm the new commit is there.
