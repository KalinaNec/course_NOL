# Introduction to python data types

### 1. Integer
Integers are whole numbers, that is, numbers without a decimal point. They form one of the basic numeric types in Python and are used whenever the represented quantity is discrete.

Typical examples are counts of animals, cells, trials, or recordings. Standard arithmetic operations such as addition, subtraction, multiplication, and division can be applied to them.

```python
print("Type of 1 is: type(1)")

print("We can add: 120 + 2 =", 120 + 2)

print("We can subtract: 120 - 2 =", 120 - 2)

print("We can multiply: 120 * 2 =", 120 * 2)

print("We can divide: 120 * 2 =", 120 / 2)

print("But type of 120/2 is", type(120 / 2))

print("So we perform floor division: 120 // 2 =", 120 // 2)

print("We can exponentiate: 120 * 2 =", 120**2)

print("We can root: 120 ** 0.5 =", 120**0.5)
```

### 2. Float

Same rules apply to floats - numerical type with a decimal point. Because computers can't represent something with infinite precision, floats have precision up to a number of bits (floating point precision). 

This can sometimes lead to missmatches between values that should give the same result but were performed by different libraries and have different internal representation of the float.

```python
print("Type of 1 is:", type(2.0))

print("We can add: 120.0 + 2.0 =", 120.0 + 2.0)

print("We can subtract: 120.0 - 2.0 =", 120.0 - 2.0)

print("We can multiply: 120.0 * 2.0 =", 120.0 * 2.0)

print("We can divide: 120.0 * 2.0 =", 120.0 / 2.0)

print("What do we get with floor division of floats? 120.0 // 2.0 =", type(20.0 // 2.0))

print("We can exponentiate: 120.0 * 2.0 =", 120.0**2.0)

print("We can root: 120.0 ** 0.5 =", 120.0**0.5)

print("We can round floats:", round(120.0**0.5))
```

### 3. String

A string is a sequence of characters used to represent text. In practice, strings are used for names, labels, sample identifiers, file paths, and other textual information.

Strings support operations and methods that are meaningful for text, such as concatenation, splitting, replacement, and changes of letter case. A string that contains digits is still text, not a numeric value.


```python
my_string = "this is a string"

print(type(my_string))

print("We can add strings:", "string_1" + "string_2")

print("We can multiply strings:", "string_1" * 10)

print("Strings can be capitalized:", my_string.capitalize())

print("Strings can be made upper case:", my_string.upper())

print("Strings can be made lower case:", my_string.lower())

print(
    "Parts of the string can be counted. How many times is 'is' present?",
    my_string.count("is"),
)

print("Parts of the string can be replaced:", my_string.replace(" ", "_"))

print("Strings can be split:", my_string.split(" "))

TRANSCRIPTION_TABLE = str.maketrans(
    {  # Creates a table with translation of ascii code of a character
        "T": "A",
        "C": "G",
        "A": "U",
        "G": "C",
    }
)

print(
    "We can even make complementary mRNA strands. For CTGCTAGGCCTCCGTGCCCGGAGTTTCGGC it will be:",
    "CTGCTAGGCCTCCGTGCCCGGAGTTTCGGC".translate(TRANSCRIPTION_TABLE),
)
```

### 4. List

A list is an ordered container type. It is one of the most frequently used Python objects because many problems involve storing and processing a sequence of items.

Lists are mutable, which means their content can be changed after creation. Elements can be added, removed, reordered, or accessed by position.


```python
my_list = ["item_1", "item_2"]
print(my_list)

my_list.append("item_3")
print(my_list)

my_list.extend(["item_4", "item_5"])
print(my_list)

print("extend method is the same as:", my_list + ["yet_another_item"], "\n")

print("or\n")

print(sum([my_list, ["yet_another_item"]], []), "\n")

print("We can multiply lists:", my_list * 3)

print("We can pop stuff out of the list:", my_list.pop(-1))

print("We can remove stuff out of the list:")
my_list.remove("item_1")
print(my_list)

print("We can put stuff into the list anywhere we want:")
my_list.insert(2, "item_1")
print(my_list)

print("We can reverse the list:")
my_list.reverse()
print(my_list)

print("We can sort the list:")
my_list.sort()
print(my_list)

print("We can get the index of an item within the list:", my_list.index("item_3"))

print("We can empty the list:", my_list.clear())
```

### 5. Tuples

A tuple is an ordered container type similar to a list. The important distinction is that tuples are immutable, so their content is not meant to be altered after creation.

Tuples are therefore well suited for fixed pieces of information, such as coordinates, dimensions, or other grouped values that conceptually belong together.


```python
my_tuple = ("item_1", "item_2")

print("We can add tuples:", my_tuple + my_tuple)

print("We can multiple tuples:", my_tuple * 10)

print("We can get the index of an item within the tuple:", my_tuple.index("item_2"))

print("We can count occurences of an item in a tuple:", my_tuple.count("item_2"))

print("Why so little?")  # Discuss mutability
```

### 6. Dictionary

A dictionary stores data as key-value pairs. Instead of accessing information by numerical position, values are retrieved through their associated keys.

This makes dictionaries especially useful when data has a natural labeling structure, for example `animal_id -> weight` or `brain_region -> mean_activity`.


```python
my_dict = {
    "name": "Konrad",
    "age": 31,
    "affiliation": "Nencki Insititute",
    "projects": [
        "Nencki Open Lab",
        "DeepEcoHab",
        "PlotlyBrain",
        "DeepLabCut",
    ],
}

print("We can check the keys:", my_dict.keys())

print("We can check the values:", my_dict.values())

print("We can check the items:", my_dict.items())


print("We can get values associated with keys:", my_dict.get("name"))

print(
    "We can pop items associated with keys and return their value:", my_dict.pop("name")
)
print(my_dict)

print("We can update values:")
my_dict.update({"affiliation": "Unemployed"})
print(my_dict)
```

### 7. Sets

A set is a container type that stores unique elements only. Unlike lists and tuples, sets are not primarily concerned with order, but with membership and uniqueness.

Sets are useful for operations such as testing whether an element is present, finding overlaps between collections, and removing duplicates.


```python
brain_regions = {
    "occipital_cortex",
    "parietal_cortex",
    "prefrontal_cortex",
    "thalamus",
    "hippocampus",
    "amygdala",
}
cortices = {"occipital_cortex", "prefrontal_cortex", "parietal_cortex"}

print(
    "We can check if cortices are a subset of brain_regions:", brain_regions >= cortices
)

print("We can get the union of sets:", brain_regions | cortices)

print("We can get the intersection of sets:", brain_regions & cortices)

print("We can get the difference between sets:", brain_regions - cortices)

print("We can add to sets:")
cortices.add("barrel_cortex")
print(cortices)

print("We can get the symmetric difference between sets:", brain_regions ^ cortices)
```

### 8. Data types as containers for information:

A useful way to think about data types is that they are different kinds of containers for information.

- `int` is for whole numbers, like number of mice in a cage.
- `float` is for numbers with a decimal point, like animal weight in grams.
- `string` is for text, like sample name or brain region name.
- `list` is for an ordered collection, like a list of trials.
- `tuple` is similar to a list, but meant to stay unchanged.
- `dict` is for matching keys to values, like `"animal_id" : body_weight`.
- `set` is for unique values only, like a collection of unique gene names.

In simple words: a data type tells Python what kind of thing it is looking at, and what operations make sense for that thing.


```python
number_of_animals = 12
mouse_weight = 24.7
brain_region = "hippocampus"
trial_names = ["trial_1", "trial_2", "trial_3"]
image_shape = (512, 512)
mouse_info = {"mouse_01": 24.7, "mouse_02": 26.1}
genes = {"fos", "gad1", "fos", "slc17a7"}

print("Integer example:", number_of_animals)
print("Float example:", mouse_weight)
print("String example:", brain_region)
print("List example:", trial_names)
print("Tuple example:", image_shape)
print("Dictionary example:", mouse_info)
print("Set example (notice unique values only):", genes)
```

### 11. Common beginner confusion

Participants often mix up the question "what value is stored?" with the question "what type of object is this?"

Examples:

- `12` and `12.0` look similar, but one is an integer and the other is a float,
- `"12"` looks like a number, but it is text,
- `[1, 2, 3]` is not one number, it is a list containing three numbers.

This is worth saying out loud during class, because many errors in beginner code come from using the right-looking value in the wrong type.


### 12. A good way to compare the container types

When introducing container types, a short practical comparison often helps:

- use a **list** when order matters and the content may change,
- use a **tuple** when order matters but the object should stay fixed,
- use a **dictionary** when you want to look up values by a key,
- use a **set** when you only care about unique values.
